# Label Encoding

```python
columns_to_encode = ['brand', 'OS', 'graphics', 'graphics_coprocessor']

# Initialize the LabelEncoder
label_encoder = LabelEncoder()

# Apply label encoding to the selected columns
for column in columns_to_encode:
    df_cleaned[column + '_encoded'] = label_encoder.fit_transform(df_cleaned[column])

# Drop the original categorical columns if needed
df_cleaned = df_cleaned.drop(columns=columns_to_encode)
```
